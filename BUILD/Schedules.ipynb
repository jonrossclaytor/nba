```python
import urllib, sqlite3
from bs4 import BeautifulSoup


# create the database
con = sqlite3.connect('C:\\NBA\\nba.db') # create database in the directory where the .py file is saved
cur = con.cursor()
cur2 = con.cursor()



cur.execute('DROP TABLE IF EXISTS schedules')
cur.execute('''CREATE TABLE schedules
             (
              team TEXT
             ,game_date TEXT
             ,opponent TEXT
             ,home_flag INTEGER
             )''')

url = 'http://www.cbssports.com/nba/schedules/team'
html = urllib.urlopen(url).read()
soup = BeautifulSoup(html, "lxml")

team_schedules = []
for a in soup.findAll('a'):
    if a.text == 'Viewable':
        team_schedules.append('http://www.cbssports.com' + a.get('href', None))
        
for team in team_schedules:
    this_team = team.split('/')
    this_team = this_team[len(this_team)-1]
    
    if this_team == 'ATL': this_team = 'Atlanta Hawks'
    elif this_team == 'BOS': this_team = 'Boston Celtics'
    elif this_team == 'BKN': this_team = 'Brooklyn Nets'
    elif this_team == 'CHA': this_team = 'Charlotte Bobcats'
    elif this_team == 'CHI': this_team = 'Chicago Bulls'
    elif this_team == 'CLE': this_team = 'Cleveland Cavaliers'
    elif this_team == 'DAL': this_team = 'Dallas Mavericks'
    elif this_team == 'DEN': this_team = 'Denver Nuggets'
    elif this_team == 'DET': this_team = 'Detroit Pistons'
    elif this_team == 'GS': this_team = 'Golden State Warriors'
    elif this_team == 'HOU': this_team = 'Houston Rockets'
    elif this_team == 'IND': this_team = 'Indiana Pacers'
    elif this_team == 'LAC': this_team = 'Los Angeles Clippers'
    elif this_team == 'LAL': this_team = 'Los Angeles Lakers'
    elif this_team == 'MEM': this_team = 'Memphis Grizzlies'
    elif this_team == 'MIA': this_team = 'Miami Heat'
    elif this_team == 'MIL': this_team = 'Milwaukee Bucks'
    elif this_team == 'MIN': this_team = 'Minnesota Timberwolves'
    elif this_team == 'NO': this_team = 'New Orleans Hornets'
    elif this_team == 'NY': this_team = 'New York Knicks'
    elif this_team == 'OKC': this_team = 'Oklahoma City Thunder'
    elif this_team == 'ORL': this_team = 'Orlando Magic'
    elif this_team == 'PHI': this_team = 'Philadelphia 76ers'
    elif this_team == 'PHO': this_team = 'Phoenix Suns'
    elif this_team == 'POR': this_team = 'Portland Trail Blazers'
    elif this_team == 'SA': this_team = 'San Antonio Spurs'
    elif this_team == 'SAC': this_team = 'Sacramento Kings'
    elif this_team == 'TOR': this_team = 'Toronto Raptors'
    elif this_team == 'UTA': this_team = 'Utah Jazz'
    elif this_team == 'WAS': this_team = 'Washington Wizards'

    html = urllib.urlopen(team).read()
    soup = BeautifulSoup(html, "lxml")
    for tr in soup('table')[3].findAll('tr'):
        if tr.get('class',None)[0] in ['row1','row2']:
            for td in tr.findAll('td'):
                if td.a != None:
                    if 'scoreboard' in td.a.get('href', None):
                        dt = td.a.get('href', None)
                        dt = dt.split('/')
                        dt = dt[len(dt)-1]
                    elif 'teams' in td.a.get('href', None):
                        opponent = td.string
                        if opponent[0] == '@':
                            home = 0
                            opponent = opponent[1:]
                        else:
                            home = 1
                            
            if opponent == 'Atl': opponent = 'Atlanta Hawks'
            elif opponent == 'Bos': opponent = 'Boston Celtics'
            elif opponent == 'Brook': opponent = 'Brooklyn Nets'
            elif opponent == 'Char': opponent = 'Charlotte Bobcats'
            elif opponent == 'Chi': opponent = 'Chicago Bulls'
            elif opponent == 'Cle': opponent = 'Cleveland Cavaliers'
            elif opponent == 'Dal': opponent = 'Dallas Mavericks'
            elif opponent == 'Den': opponent = 'Denver Nuggets'
            elif opponent == 'Det': opponent = 'Detroit Pistons'
            elif opponent == 'GS': opponent = 'Golden State Warriors'
            elif opponent == 'Hou': opponent = 'Houston Rockets'
            elif opponent == 'Ind': opponent = 'Indiana Pacers'
            elif opponent == 'LAC': opponent = 'Los Angeles Clippers'
            elif opponent == 'LAL': opponent = 'Los Angeles Lakers'
            elif opponent == 'Mem': opponent = 'Memphis Grizzlies'
            elif opponent == 'Miami': opponent = 'Miami Heat'
            elif opponent == 'Mil': opponent = 'Milwaukee Bucks'
            elif opponent == 'Min': opponent = 'Minnesota Timberwolves'
            elif opponent == 'NO': opponent = 'New Orleans Hornets'
            elif opponent == 'NY': opponent = 'New York Knicks'
            elif opponent == 'OKC': opponent = 'Oklahoma City Thunder'
            elif opponent == 'Orl': opponent = 'Orlando Magic'
            elif opponent == 'Phi': opponent = 'Philadelphia 76ers'
            elif opponent == 'Pho': opponent = 'Phoenix Suns'
            elif opponent == 'Port': opponent = 'Portland Trail Blazers'
            elif opponent == 'SA': opponent = 'San Antonio Spurs'
            elif opponent == 'Sac': opponent = 'Sacramento Kings'
            elif opponent == 'Tor': opponent = 'Toronto Raptors'
            elif opponent == 'Utah': opponent = 'Utah Jazz'
            elif opponent == 'Wash': opponent = 'Washington Wizards'
            
            dt = dt[:4] + '-' + dt[4:6] + '-' + dt[-2:]
            
            sched = (this_team, dt, opponent, home)
            cur2.execute('INSERT INTO schedules VALUES(?' + ',?'*3 + ')',sched)
            con.commit()
con.close()
                    
```
