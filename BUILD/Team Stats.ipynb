```python
import sqlite3

con = sqlite3.connect('C:\\NBA\\nba.db')
cur = con.cursor()
cur2 = con.cursor()
cur3 = con.cursor()
cur4 = con.cursor()


cur.execute('DROP TABLE IF EXISTS team_stats')
cur.execute('''CREATE TABLE team_stats
             (
                 team TEXT
                ,game_id INTEGER
                ,season INTEGER
                ,game_date TEXT
                ,missed_field_goal_pg_season REAL
                ,missed_three_pointers_pg_season REAL
                ,missed_free_throws_pg_season REAL
                ,offensive_rebounds_pg_season REAL
                ,defensive_rebounds_pg_season REAL
                ,assists_pg_season REAL
                ,steals_pg_season REAL
                ,blocks_pg_season REAL
                ,turnovers_pg_season REAL
                ,personal_fouls_pg_season REAL
                ,missed_field_goal_pg_5games REAL
                ,missed_three_pointers_pg_5games REAL
                ,missed_free_throws_pg_5games REAL
                ,offensive_rebounds_pg_5games REAL
                ,defensive_rebounds_pg_5games REAL
                ,assists_pg_5games REAL
                ,steals_pg_5games REAL
                ,blocks_pg_5games REAL
                ,turnovers_pg_5games REAL
                ,personal_fouls_pg_5games REAL
                ,missed_field_goal_pg_twoweeks REAL
                ,missed_three_pointers_pg_twoweeks REAL
                ,missed_free_throws_pg_twoweeks REAL
                ,offensive_rebounds_pg_twoweeks REAL
                ,defensive_rebounds_pg_twoweeks REAL
                ,assists_pg_twoweeks REAL
                ,steals_pg_twoweeks REAL
                ,blocks_pg_twoweeks REAL
                ,turnovers_pg_twoweeks REAL
                ,personal_fouls_pg_twoweeks REAL
             )''')


cur.execute("SELECT DISTINCT team FROM games")# WHERE team = 'Memphis Grizzlies'")

for team in cur:
    cur2.execute("SELECT DISTINCT game_id, game_date, season FROM games WHERE team = '" + team[0] + "'")
    for game in cur2:
        game_id = str(game[0])
        game_date = str(game[1])
        season = str(game[2])
        query = open('C:\\NBA\\SQL\\team_template.sql','r')
        query = query.read()

        query_new = query.replace('<team>',team[0])
        query_new = query_new.replace('<season>',season)
        query_new = query_new.replace('<game_date>',str(game_date))
        cur3.execute(query_new)
        game_insert = [team[0], game_id, season, game_date]
        for game in cur3:
            game = list(game)
            for item in game:
                game_insert.append(item)
            #print game_insert
            cur4.execute('INSERT INTO team_stats VALUES(?' + ',?'*33 + ')',game_insert)
    con.commit()    # commit inserted rows
con.close()
            
```
